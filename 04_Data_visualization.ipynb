## 1-  Line Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns

# load data from library seaborn
data= sns.load_dataset('iris')

# Setting figure size and style of plot
plt.figure(figsize=(8,4))
plt.style.use('ggplot') # ggplot, seaborn, fivethirtyeight, bmh, dark_background, grayscale, classic , default

# DRAW LINE PLOT
sns.lineplot(x="sepal_length", y="sepal_width", data=data, color='red',
             label='sepal', marker='o', markersize=5, linewidth=2, linestyle='dashed', alpha=1)

# color : red, green, blue, cyan, magenta, yellow, black, white
# label : label of line plot (legend) 
# marker : marker of line plot (o, v, ^, <, >, s, p, *, h, H, D, d, P, X)
# markersize : size of marker 
# linewidth : width of line 
# linestyle : style of line plot (-, --, -., :, None, '', ' ', 'solid', 'dashed', 'dashdot', 'dotted')
# alpha : transparency of line plot (0 to 1)

# Adding titles 
plt.title("comparision between sepal length and sepal width")
plt.xlabel("sepal length")
plt.ylabel("sepal width")

# Setting limit of X-axis and Y-axis
plt.xlim(3,9)
plt.ylim(2, 4)

# show plot
plt.show()
```

## 2-  bar plot

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Loading dataset from library seaborn
data = sns.load_dataset("titanic")

# Setting figure size and style of plot
plt. figure(figsize=(8,4))
plt. style.use('ggplot') # ggplot, seaborn, fivethirtyeight, bmh, dark_background, grayscale, classic , default

# Adding titles and labels 
plt.title("bar plot shows survived with ages")
plt.xlabel("Survived")
plt.ylabel("Age")

# DRAW BAR PLOT
sns.barplot (x="sex", y="age", hue="who",data= data,  palette="Set2",
            ci=99, capsize=0.3, dodge=True, saturation=1, errcolor="0.1", 
            errwidth=2, alpha=1, linewidth=2, edgecolor="0.1", orient="v")
# x : x-axis data
# y : y-axis data
# hue : group data by hue 
# palette : color of plot (Set1, Set2, Set3, deep, muted, pastel, bright, dark, colorblind)
# ci : confidence interval (95, 99, None) # None means no error bar
# capsize : cap size of error bar
# dodge : dodge the bars at the categorical locations (True, False)
# saturation : saturation of the color (0 to 1)  # 0 means gray and 1 means full color 
# errcolor : color of error bar (0 to 1) or (red, green, blue, cyan, magenta, yellow, black, white)
# errwidth : width of error bar  
# alpha : transparency of plot (0 to 1) 
# linewidth : width of line 
# edgecolor : color of edge (0 to 1) or (red, green, blue, cyan, magenta, yellow, black, white)
# orient : orientation of plot (v, h) # v means vertical and h means horizontal

# show plot
plt.show()
```

## 3- Box plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Import dataset
tip = sns.load_dataset("tips")

# Setting figure size and style of plot
plt. figure(figsize=(8,4))
plt.style.use('ggplot') # ggplot, seaborn, fivethirtyeight, bmh, dark_background, grayscale, classic , default

# DRAW BOX PLOT 
sns.boxplot(x="day", y="tip", 
             hue="smoker",
             palette="Set2",
             dodge=True,
             data=tip)

# x : x-axis data
# y : y-axis data
# hue : group data by hue 
# palette : color of plot (Set1, Set2, Set3, deep, muted, pastel, bright, dark, colorblind)
# dodge : dodge the bars at the categorical locations (True, False) catagories overlap krny k liye

# Adding titles and labels X-axis and Y-axis
plt.title("box plot shows tips with days", fontsize=20, color="blue", fontweight="bold",
          fontstyle="italic", loc="center", verticalalignment="baseline", horizontalalignment="center")
plt.xlabel("DAYS", fontsize=20, color="pink", fontweight="bold")
plt.ylabel("TIP", fontsize=20, color="pink", fontweight="bold")

# show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Import dataset
data = sns.load_dataset("titanic")

# DRAW BOX PLOT
#showmeans = True means show mean of data in box plot otherwise False
# meanproprs = mean properties of box plot (marker, markerfacecolor, markeredgecolor, markersize)

sns.set_style(style="whitegrid")
sns.boxplot(x="survived", y="age",
            showmeans=True,
            meanprops={"marker": "o",
                        "markersize": "8",
                        "markeredgecolor": "black"},
                         data=data)

# Adding titles and labels X-axis and Y-axis
plt.title("Box plot Survivals with ages", size=15, weight="bold")
plt.xlabel("Survivals", size=14)
plt.ylabel("Age", size=14)

# FACET GRID BOX PLOT 
g = sns.FacetGrid(data, col="survived",  row="sex")
g.map(sns.boxplot, "class", "age")

# col : column name
# row : row name
# FacetGrid is used to draw multiple plots on a single figure
```
