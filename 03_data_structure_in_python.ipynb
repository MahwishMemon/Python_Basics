#### 1- Tuple
- tuple is a collection which is ordered and unchangeable once it create. Allows duplicate members.
- tuple is enclosed in round brackets (). in tuple we can store different data types.

```python
# tuple1 contains different data types lie int, float, string, list, tuple, dictionary  

tup1 = (55, "love-life", True, 2.5)
print(tup1)

# Concatination of tuples  
print(tup1 + tup1)

# Repitition of tuples like here 3 times
print(tup1*6)

# Length of a tuple 
print("length of tuple 1  is  : ",len(tup1)) # length of tuple 1 is 4  

# Min and Max of a tuple2
tup2 = (1,2,3,4,5,6,7,8,9)
print("minimum of this tuple 2 is :", min(tup2),"maximum of this tuple 2 is :", max(tup2))

# Indexing
print(tup2[1:4])

# Counts the number of times an element occured 
tup3 = (1,1,2,2,3,3,4,4,4,4)
print(tup3.count(3))

# Tells position of the element 4 in the tuple
print(tup3.index(4))
```

### 2- List
* List is a collection which is ordered and changeable. Allows duplicate members.
* list is enclosed in square brackets [ ].in list we can store different data types.


```python
# make a list of tuples 
list1 = ["Me", 55, False]
list2 = ["food", 13, True]
list3 = [3,5,8,5,0,9,5,7,44]

# Concatination
print(list1 + list2)

# Repitition
print(list1*2)

# Add an element in list
list3.append(100)
print(list3)

# Remove all elements from list
list3.clear()
print(list3)

# Create a new copy of list
list4 = list3.copy()
list4.append(8) # add 8 in list4
print(list4)

# Count the number of times an element occured in a list
print(list4.count(8))

# Insert another list in the existing list
list4.extend(list1)
print(list4)

# Find the position of an element in the list
print(list4.index("Me"))

# Add an element in list in a certain Index
list4.insert(9, 9)
print(list4)

# Remove and return an element based on index
pop = list4.pop(1)
print(list4)

# Remove an element from the list
list4.remove(False)
print(list4)

# Reverse the positions of all elements in the list
list4.reverse()
print(list4)

# Sort the list in asc/des order
list5 = [5,7,2,1,3,8]
list5.sort()
print(list5)
```

### 3- Dictionary 
- Dictionary is a collection of key-value pairs.
- Dictionary is a mutable and unordered data type.
- Dictionary is a data type that is written with curly brackets, and has keys and values.
- Dictionary is a data type that is used to store data values in key:value

```python
# make a dictionary of clothes and their prices
cloths_set1= {"shirt": 1000, "dupata":200, "trouser": 2000, "shoes": 3000, "belt": 4000, "watch": 5000}
cloths_set2= {"shirt": 5000, "trouser": 300, "shoes": 1500, "belt": 200, "watch": 10000}
print("This is cloth set1  ", cloths_set1)

# Checking the type of a variable in cloths_set1
print("type of cloth_set1 is :", type(cloths_set1))

# Add a new key:value in dict or cloths_set1
cloths_set1["cap"] = 100
print("this is updated cloth_set1", cloths_set1)

# Modfiying the value of a key in dict or cloths_set1
cloths_set1["trouser"] = 770
print("this is updated cloth_set1",cloths_set1)

# Add another dict into the already existing dict
cloths_set1.update(cloths_set2)
print("cloths_set1 is updated with cloths_set2",cloths_set1)
print("this is cloths_set2", cloths_set2)

# Remove all keys:values from a dict
cloths_set2.clear()
print("this is cloths_set2 after clearing", cloths_set2)

# Creating a new copy of the dict or cloths_set3
cloths_set3 = cloths_set1.copy()
print("this is copy of cloth set 2", cloths_set3)


# Get value from a dict using the key
print ("the value of 'shirt' in cloth set1 is :", cloths_set1.get("shirt"))
# Get all key:value pairs of a dict
print(cloths_set1.items())
# Get all keys from a dict
print(cloths_set1.keys())
# Get VALUES of a dict or cloths_set1
print(cloths_set1.values())

# Remove and return a value of a key
pop = cloths_set1.pop("cap")
print(pop)
```

### 4- Set
- A set is a collection which is unordered and unindexed. In Python sets are written with curly brackets.
- due to unordered, so you cannot be sure in which order the items will appear.
- Sets are unchangeable, meaning that we cannot change the items after the set has been created.
- Sets do not allow duplicate values.

```python
# make a set of numbers 

set1 = {1,2,3.5,55.9, 77, "Ahmed", True}
print("this is set 1 ",set1)

# Adding an element to a set1
set1.add("Abbas")
print("Updated set 1", set1)


# Creates a new copy of the set
set2 = set1.copy()
print("its set 2 which is copy of updated set 1",set2)

# Adding an element to a set2 
set2.add( "Areeba")
print("Updated set 2", set2)

# Finds the diff of the two sets
print("Differance between set1 and set2 is ", set2.difference(set1))

# Sets the value of the diff as the only value of the first set 
set1.difference_update(set2)
print("set 1 is updated with the differance of set1 and set2", set1, set2)

# Removes the given element
set2.discard('Abbas')
print("discard a variable from set2 ", set2)

# Finds the intersection of the two sets set2 and set 3
set3 = {"Hello", 1, 2, 3.5, 4, 5, 6, 77, 8, 9, 10}
print("this is set 3",set3)
print("common value between set3 and set2",set2.intersection(set3)) 

# Checks if the two sets contain not a single similiar element 
print(set2.isdisjoint(set3)) 
# Returns true if the set is empty or returns false if the set is not empty
print(set2.isdisjoint(set1))

# Checks if first set is a subset of second set
print(set1.issubset(set2)) 
# returns false if set1 is not a subset of set2 
# returns true if set1 is a subset of set2


# Removes the element and returns it. We can store that removed element in another variable
set2 = {1, 2, 3.5, 55.9, 'Areeba', 'Ahmed', 77}
set2.pop() 
print(set2)

# Removes the elements completely
set2.remove(3.5)
print(set2)

# Finds the sym-diff of the two sets
sem_set = set2.symmetric_difference(set1)
print(sem_set)

# Merges the two groups
set_a = {1,2,3}
set_b = {4,5,6}
print(set_a.union(set_b))

# Joins the two groups
set_a.update(set_b)
print(set_a)


# Removes all the elements
set1.clear()
print("set 1 is cleared", set1)
```
